```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd

breastData = load_breast_cancer()
breastData
```

```python
breastData.data
```

```python
breastData.target
```

```python
# reshape target to 2D
labels = np.reshape(breastData.target, (569, 1))
```

```python
labels
```

```python
breast_data = np.concatenate([breastData.data, labels], axis=1)
```

```python
breast_data
```

```python
breast_data[:,30]
```

```python
breastDF = pd.DataFrame(breast_data)
breastDF
```

```python
# menambahkan kolom

breastFeatureLabel = np.append(breastData.feature_names, "label")
```

```python
breastDF.columns = breastFeatureLabel
```

```python
breastDF
```

```python
breastDF["label_name"] = np.where(breastDF["label"] == 0, "Benign", "Malignant")
```

```python
breastDF
```

Normalize

```python
from sklearn.preprocessing import StandardScaler
x = breastDF.loc[:, breastData.feature_names].values
x = StandardScaler().fit_transform(x)
x
```

```python
# menentukan nama kolom

column_name = ["feature"+str(i) for i in range(x.shape[1])]
breastNormalize = pd.DataFrame(x, columns=column_name)
```

```python
breastNormalize
```

PCA

```python
from sklearn.decomposition import PCA

pca_model = PCA(n_components=2)
principal_com_breast = pca_model.fit_transform(x)
```

```python
principal_com_breast
```

```python
pca_model.explained_variance_ratio_ # memberikan berapa persen informasi
```

```python
pca_breast_df = pd.DataFrame(principal_com_breast, columns=["Principal 1", "Principal 2"])
```

```python
pca_breast_df
```

Visualisasi

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.figure(figsize=(12,7))
plt.xlabel("Principal Component -1")
plt.ylabel("Principal Component -2")
plt.title("Principal Component Breast Cancer")

targets = ["Benign", "Malignant"]
colors = ["r", "g"]

for target, color in zip(targets, colors):
    idxToKeep = breastDF["label_name"] == target
    plt.scatter(pca_breast_df.loc[idxToKeep, "Principal 1"],
            pca_breast_df.loc[idxToKeep, "Principal 2"],
            c=color, s=40)
plt.legend(targets)
```
