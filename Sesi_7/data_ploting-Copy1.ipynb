```python
continent_Data = df_can.groupby("Continent").sum()
```

```python
color_list = ["Purple", "yellow", "orange", "Black", "Blue", "Pink", "gray"]
explode_list = [0.1,0,0,0,0,0.1,0.1]
continenttotalPlot = continent_Data["Total"].plot(kind="pie", 
                                                  figsize = (15,5), 
                                                  colors=color_list,
                                                  explode = explode_list,
                                                  autopct="%1.1f%%",
                                                  shadow=True,
                                                  labels=None,
                                                  pctdistance=1.13,
                                                  startangle=90
                                                 )
plt.title("Immigritaion to Canada by Continent[1980 - 2013]", y=1.12)

plt.axis("equal")

#add legend
plt.legend(labels=continent_Data.index, loc= "upper left")
```
