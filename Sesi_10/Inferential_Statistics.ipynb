## Inferential Statistics (why we need)
* Mengambil kesimpulan dari populasi berdasarkan sample
* membandungkan sample dari populasi apakah sample kita benar-benar berbeda dari populasi
* model machine learning - justifikasi menambahkan atau mengurangi ciri
* menguji hipotesa.


### Random variable: variable yang kita amati dan nilainya dari fenomena acak
* discreate: 1,2,3,4,5,6 (ex: jumlah orang, jumlah kendaraan, total pemudik) (curva disebut mass curve)
* continous: 1.24, 1.26 (ex: pendapatan, curah hujan, tinggi badan) (curva disebut dencity curve)

```python
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
```

```python
from scipy.stats import uniform

n = 10000
start = 10
width = 20

dat_uniform = uniform.rvs(size=n, loc=start, scale=width)

dat_uniform
```

```python
data_plot = sns.distplot(dat_uniform, bins=20, color="red")
data_plot.set(xlabel = "Uniform DIS", ylabel = "Frequency")
```

```python
# normal Distribution 
from scipy.stats import norm

n = 10000
start = 10
width = 20
fig = plt.figure(figsize=(16, 8))
dat_norm = norm.rvs(size= n, loc = start, scale=width) 
                 
norm_plot = bins = sns.distplot(dat_norm, 
                                bins=20,
                                color="purple", 
                                kde=True) 

norm_plot.hist(dat_norm, density=True, histtype='stepfilled', alpha=0.5)
norm_plot.set(xlabel = "Normal Distribution", ylabel="Frequency")
```

```python
# gamma distribution chi-squared

from scipy.stats import gamma

dat_gamma = gamma.rvs(size=n, a=10)
dat_gamma = sns.distplot(dat_gamma, bins=20, color="red")
```

```python
dat_gamma = gamma.rvs(size=n, a=1)
dat_gamma = sns.distplot(dat_gamma, bins=50, color="red")
```

```python
# exponential

from scipy.stats import expon
dat_expon = expon.rvs(size=n, scale=2, loc=0)
plot_expon = sns.distplot(dat_expon, bins=50, color="blue")
```

```python
# Poisson Distribution 

from scipy.stats import poisson
dat_poisson = poisson.rvs(size=n, mu=5)
plot_poisson= sns.distplot(dat_poisson, bins=50, color="blue")
```
