```python
train.fillna(train.mean(), inplace=True)
train.isnull().sum()
```

```python
train.fillna(train.Gender.mode()[0], inplace=True)
train.fillna(train.Married.mode()[0], inplace=True)
train.fillna(train.Dependents.mode()[0], inplace=True)
train.fillna(train.Self_Employed.mode()[0], inplace=True)
train.isnull().sum()
```
